```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://zoro.to/movie'
```

```python
#Headers for request
#gt user-agent from whatismybrowser
HEADERS = ({'User-Agent':'',    'Accept-Language':'en-US, en;q=0.5'})
```

```python
#Http Request
webpage = requests.get(url, headers = HEADERS)
webpage
```

```python
#to show entire html document
webpage.content
```

```python
type(webpage.content)
```

```python
#convert webpage content into html format
soup = BeautifulSoup(webpage.content, 'html.parser')
```

```python
soup
```

```python
#Fetch links as List of Tag Objects
links = soup.find_all('a',attrs={'class': 'dynamic-name'})
```

```python
links
```

```python
links[0]
```

```python
link = links[1].get('href')
```

```python
link
```

```python
product_list = "https://zoro.to" + link
```

```python
product_list
```

```python
#getting the html content form our new link
new_webpage = requests.get(product_list, headers=HEADERS)
```

```python
new_webpage
```

```python
#using soup object to parse in proper html structure. the soup object contains all data
new_soup = BeautifulSoup(new_webpage.content, "html.parser")
```

```python
new_soup
```

```python
anime_title = soup.find('a',attrs={'class': 'dynamic-name'}).text.strip()
anime_title
```

```python
#finding all titles but title is also surrounded by the link tag
anime_info = new_soup.find_all('div',attrs={'class': 'item item-title'})
anime_info
```

```python
#find all anime infor in span tag
anime_info_inspan = new_soup.find_all('span',attrs={'class': 'name'})
anime_info_inspan
```

```python
for anime in anime_info_inspan:
    print(anime.text)
```

```python
anime_info_inspan[0].text
```

```python
anime_dict ={'AnimeTitle':[],'Japanese':[],'Aired':[], 'Premiered':[], 'Duration':[], 'Status':[], 'MAL Score':[],
            'Studios':[], 'Producers':[]}
```

```python
anime_dict['AnimeTitle'].append(anime_title)
anime_dict['Japanese'].append(anime_info_inspan[0].text)
anime_dict['Aired'].append(anime_info_inspan[1].text)
anime_dict['Premiered'].append(anime_info_inspan[2].text)
anime_dict['Duration'].append(anime_info_inspan[3].text)
anime_dict['Status'].append(anime_info_inspan[4].text)
anime_dict['MAL Score'].append(anime_info_inspan[5].text)
```

```python
anime_dict
```

```python
anime_info[-1].find('a', attrs={'class':'name'}).text.strip()
anime_info[-2].find('a', attrs={'class':'name'}).text.strip()
```

```python
anime_dict['Studios'].append(anime_info[-2].find('a', attrs={'class':'name'}).text.strip())
anime_dict['Producers'].append(anime_info[-1].find('a', attrs={'class':'name'}).text.strip()
)
```

```python
anime_dict
```

```python
anime_dataframe = pd.DataFrame.from_dict(anime_dict)
```
