```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
#Function to Extract Anime Title
def get_title(soup):
    try:
        title = soup.find('a',attrs={'class': 'dynamic-name'}).text.strip()
    except AttributeError:
        title = ""
    return title
                          
def get_studioInfo(soup):
    studios = ''
    try:
        anime_prodinfo = soup.find_all('div',attrs={'class': 'item item-title'})
        studios = anime_prodinfo[-2].find('a', attrs={'class':'name'}).text.strip()
    except AttributeError:
        studios = ""
    return studios

def get_productionInfo(soup):
    production = ''
    try:
        anime_prodinfo = soup.find_all('div',attrs={'class': 'item item-title'})
        production = anime_prodinfo[-1].find('a', attrs={'class':'name'}).text.strip()
    except AttributeError:
        production = ""
    return production


#Function to Extract all descriptioninfo
def get_descinfo(soup):
    try:
        anime_descinfo = soup.find_all('span',attrs={'class': 'name'})
    except:
        anime_descinfo
    return anime_descinfo
```

```python
if __name__ == '__main__':
    #add user agent
    #get user-agent from whatismybrowser
    HEADERS = ({'User-Agent':'put user agent to use','Accept-Language':'en-US, en;q=0.5'})
    
    # The webpage URL
    URL = 'https://zoro.to/movie'

    # HTTP Request
    webpage = requests.get(URL, headers=HEADERS)

    # Soup Object containing all data
    soup = BeautifulSoup(webpage.content, "html.parser")

    # Fetch links as List of Tag Objects
    links = soup.find_all('a',attrs={'class': 'dynamic-name'})

    # Store the links
    links_list = []

    # Loop for extracting links from Tag Objects
    for link in links:
            links_list.append(link.get('href'))

    anime_dict ={'AnimeTitle':[],'Japanese':[],'Aired':[], 'Premiered':[],'Synonyms':[], 'Duration':[], 'Status':[], 'MAL Score':[],
            'Studios':[], 'Producers':[]}    
    
    # Loop for extracting product details from each link 
    for link in links_list:
        new_webpage = requests.get("https://zoro.to" + link, headers=HEADERS)

        new_soup = BeautifulSoup(new_webpage.content, "html.parser")

        # Function calls to display all necessary product information
        
        #animetitle
        anime_dict['AnimeTitle'].append(get_title(new_soup))
        
        #description information
        anime_descinfo = get_descinfo(new_soup)
        if len(anime_descinfo) < 7:
#             del anime_descinfo[1]
            anime_dict['Japanese'].append(anime_descinfo[0].text)
            anime_dict['Aired'].append(anime_descinfo[1].text)
            anime_dict['Premiered'].append(anime_descinfo[2].text)
            anime_dict['Duration'].append(anime_descinfo[3].text)
            anime_dict['Status'].append(anime_descinfo[4].text)
            anime_dict['MAL Score'].append(anime_descinfo[5].text)
        else:
            anime_dict['Japanese'].append(anime_descinfo[0].text)
            anime_dict['Synonyms'].append(anime_descinfo[1].text)
            anime_dict['Aired'].append(anime_descinfo[2].text)
            anime_dict['Premiered'].append(anime_descinfo[3].text)
            anime_dict['Duration'].append(anime_descinfo[4].text)
            anime_dict['Status'].append(anime_descinfo[5].text)
            anime_dict['MAL Score'].append(anime_descinfo[6].text)
        
        #studios information
        stud_info = get_studioInfo(new_soup)
        anime_dict['Studios'].append(stud_info)
        
        #production information
        prod_info = get_productionInfo(new_soup)
        anime_dict['Producers'].append(prod_info)
```

```python
anime_df = pd.DataFrame.from_dict(anime_dict, orient='index')
```
