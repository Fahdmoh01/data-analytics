```python
'''
Qualitative Data is data that can be observed and recorded but not numerical in nature. It is also known as categorical
data.

-> Bar Charts are normally used to represent Qualitative Data

Subtypes are nominal data and ordinal data.
Nominal data: that can't be ordered
Ordinal data: is data that can be categorizeds and ordered

'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
listings = pd.read_csv('listings.csv')
```

```python
listings
```

### Creating a bar chart from neighbourhood

```python
sn.countplot(x = 'neighbourhood_group', data = listings)
plt.show()
```

```python
sn.barplot(x = 'neighbourhood_group', y='price', data = listings)
plt.show()
```

The black lines represent confidence intervals. Setting a variable ci = False removes the black lines from the barchart

```python
sn.barplot(x = 'neighbourhood_group', y='price', data = listings, ci = False)
plt.show()
```
